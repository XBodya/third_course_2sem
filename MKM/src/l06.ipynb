Составить систему Лотки-Вольтерры и найти приближенное задачи Коши для этой системы
для двух вариантов начальных условий (см. таблицу для своего варианта). Построить графики
и сделать выводы, что изменится при изменении начальных данных. Временной промежуток
студент выбирает самостоятельно, исходя из своей задачи

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def system(t, z):
    x, y = z
    dxdt = 0.42 * x - 0.05 * x * y
    dydt = 0.37 * x * y - 0.019 * y
    return [dxdt, dydt]


x0 = 103
y0 = 9
z0 = [x0, y0]

x0 = 83
y0 = 10
z0 = [x0, y0]


t_span = (1000, 3500)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

sol = solve_ivp(system, t_span, z0, t_eval=t_eval, method='RK45')

if sol.success:
    plt.plot(sol.t, sol.y[0], label='Pseudomonas putida (жертва)')
    plt.plot(sol.t, sol.y[1], label='Poteriospumella lacustris (хищник).')
    plt.xlabel('t')
    plt.ylabel('Численность популяции')
    plt.title('Численное решение системы Лотки-Вольтерры,')
    plt.legend()
    plt.ylim(0, 22)
    plt.grid()
    plt.show()
else:
    print("Решение не удалось найти")
```
