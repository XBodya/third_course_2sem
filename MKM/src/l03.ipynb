```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.animation as animation
from tabulate import tabulate
```

```python
# условие
a, b = 1, 2
N = 10
start_condition = [3, 1]
q = lambda x : np.cos(x)
f = lambda x: -np.sin(x)
h = (b - a) / N
```

```python
my_headers = [
    'i',
    'x',
    'q(x)',
    'f(x)',
    'alpha',
    'beta',
    'u_i',
    "u_i2",
    'eps'
]
```

```python
x = np.linspace(a, b, N + 1)
alpha = [0]
beta = [start_condition[0]]
u = [0 for i in range(N + 1)]
u[N] = start_condition[-1]

for i in range(1, N + 1):
    alpha.append(1 / (2 + h * h * q(x[i]) - alpha[i - 1]))

    beta.append((h * h * f(x[i]) + beta[i - 1]) * alpha[-1])

for i in range(N - 1, -1, -1):
    u[i] = alpha[i] * u[i + 1] + beta[i]

data = np.array([
    [i for i in range(N + 1)],
    [x[i] for i in range(N + 1)],
    list(map(q, x)),
    list(map(f, x)),
    alpha,
    beta,
    u
])
```

```python
print(tabulate(data.T, headers=my_headers))
```

```python
# plt.plot(x, data[2])
# plt.plot(x, data[3])
plt.plot(x, u)
plt.plot(x, u, '.', c='b')
plt.show()
```
